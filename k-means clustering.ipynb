```python
import numpy as np
import pandas as pa
from pandas import plotting
from matplotlib import pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
```

```python
d=pa.read_csv(r'C:\Users\mrambha\Downloads\Mall_Customers.csv')
d.head()
```

```python
d.describe()
```

```python
d.isna().sum()
```

```python
plt.rcParams['figure.figsize'] = (10, 5)


plotting.andrews_curves(d.drop("CustomerID", axis=1), "Gender")
plt.title('Andrew Curves for Gender', fontsize = 20,color='r')
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (15, 5)

plt.subplot(1, 2, 1)
sns.set(style = 'whitegrid')
sns.distplot(d['Annual Income (k$)'])
plt.title('Distribution of Annual Income', fontsize = 20,color="red")
plt.xlabel('Range of Annual Income')
plt.ylabel('Count',color='y')


plt.subplot(1, 2, 2)
sns.set(style = 'whitegrid')
sns.distplot(d['Age'], color = 'red')
plt.title('Distribution of Age', fontsize = 20,color="red")
plt.xlabel('Range of Age')
plt.ylabel('Count',color="b")
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15, 5)
sns.countplot(d['Age'], palette = 'rainbow')
plt.title('Distribution of Age', fontsize =20,color="brown")
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15,5)
sns.countplot(d['Annual Income (k$)'], palette = 'rainbow')
plt.title('Distribution of Annual Income', fontsize = 20,color='cyan')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15, 6)
sns.countplot(d['Spending Score (1-100)'], palette = 'copper')
plt.title('Distribution of Spending Score', fontsize = 20,color="cyan")
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10,5)
sns.heatmap(d.corr(), cmap = 'Wistia', annot = True)
plt.title('Correlation of data', fontsize = 20,color="red")
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10,5)
sns.stripplot(d['Gender'], d['Age'], palette = 'Purples', size = 5)
plt.title('Gender vs Spending Score', fontsize = 20,color="cyan")
plt.show()
```
